```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#отображение графиков прямо в ноутбуке
%matplotlib inline
```

```python
anime = pd.read_csv('anime.csv')
ratings = pd.read_csv('rating.csv')

# Проверяем размеры и первые строки
print("anime shape:", anime.shape)
anime.head()
```

```python
# убрать неоцененные записи
ratings_clean = ratings[ratings['rating'] != -1]
ratings_clean.head()
```

```python
# объединяем данные: рейтинг + инфо об аниме
df = pd.merge(ratings_clean, anime, on='anime_id', how='left')
df.shape
df.head()
```

```python
# Какие типы аниме (TV, Movie, OVA…) самые популярные?

# Считаем количество записей по каждому типу
type_counts = df['type'].value_counts()

# Строим график
plt.figure(figsize=(8,5))
sns.barplot(x=type_counts.index, y=type_counts.values)

plt.title('Типы аниме по количеству записей')
plt.xlabel('Тип')
plt.ylabel('Количество')
plt.tight_layout()
plt.show()
```
