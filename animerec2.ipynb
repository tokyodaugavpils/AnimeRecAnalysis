```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#отображение графиков прямо в ноутбуке
%matplotlib inline
```

```python
anime = pd.read_csv('anime.csv')
ratings = pd.read_csv('rating.csv')

# Проверяем размеры и первые строки
print("anime shape:", anime.shape)
anime.head()
```

```python
# убрать неоцененные записи
ratings_clean = ratings[ratings['rating'] != -1]
ratings_clean.head()
```

```python
# объединяем данные: рейтинг + инфо об аниме
df = pd.merge(ratings_clean, anime, on='anime_id', how='left')
df.shape
df.head()
```

```python
# Средний рейтинг по отдельным жанрам

# 1. Убираем строки без жанра или рейтинга
df_genres = df[['genre', 'rating_y']].dropna()

# 2. Разбиваем жанры и "взрываем"
df_genres['genre'] = df_genres['genre'].str.split(',')
df_genres = df_genres.explode('genre')
df_genres['genre'] = df_genres['genre'].str.strip()

# 3. Группируем по жанру и считаем средний рейтинг
genre_avg_rating = df_genres.groupby('genre')['rating_y'].mean().sort_values(ascending=False)

# 4. Строим график
plt.figure(figsize=(15,6))
sns.barplot(x=genre_avg_rating.index, y=genre_avg_rating.values)

# 5. Оформление графика
plt.xticks(rotation=45)
plt.title('Средний рейтинг по отдельным жанрам')
plt.xlabel('Жанр')
plt.ylabel('Средний рейтинг')
plt.tight_layout()
plt.show()
```
