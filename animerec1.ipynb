```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#отображение графиков прямо в ноутбуке
%matplotlib inline
```

```python
anime = pd.read_csv('anime.csv')
ratings = pd.read_csv('rating.csv')

# Проверяем размеры и первые строки
print("anime shape:", anime.shape)
anime.head()
```

```python
# убрать неоцененные записи
ratings_clean = ratings[ratings['rating'] != -1]
ratings_clean.head()
```

```python
# объединяем данные: рейтинг + инфо об аниме
df = pd.merge(ratings_clean, anime, on='anime_id', how='left')
df.shape
df.head()
```

```python
# этот график покажет, какие оценки ставят чаще всего.
plt.figure(figsize=(8,4))
sns.countplot(x='rating_x', data=df, palette='viridis')
plt.title('Распределение оценок пользователей')
plt.xlabel('Оценка')
plt.ylabel('Количество')
plt.show()
```
